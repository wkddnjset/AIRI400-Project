```python
def get_filenames_and_classes(dataset_dir, train_val):
    wav_root = os.path.join(dataset_dir, train_val)
    directories = []
    class_names = []
    for filename in os.listdir(wav_root):
        path = os.path.join(wav_root, filename)
        if os.path.isdir(path):
            directories.append(path)
            class_names.append(filename)
    class_name = class_names[5]
    wave_filenames = []
#     print (directories[5])
#     print (class_names)
    path = os.path.join(directories[5], filename)
#     print(os.listdir(directories[5]))
    for filename in os.listdir(directories[5]):
#         print (filename)
        path = os.path.join(directories[5], filename)
        wave_filenames.append(path)

    return wave_filenames, class_name

wave_filenames, wave_class_name = get_filenames_and_classes(_DATA_DIR, "9th Wonder Kit/")
```
