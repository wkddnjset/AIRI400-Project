## MIDI 크롤링

- requests와 BS4를 사용한 웹 크롤링
- http://www.cool-midi.com/rap-hip-hop-midi.php

```python
import requests
from bs4 import BeautifulSoup
```

```python
req = requests.get('http://www.cool-midi.com/rap-hip-hop-midi.php')
html = req.text
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
lists = soup.select('#artistlist .song a')
```

```python
artist_lists = []
for artist_list in lists:
    back = artist_list.get('href')
    full = 'http://www.cool-midi.com/'+back
    artist_lists.append(full)
```

```python
song_lists = []
for i in range(len(artist_lists)):
    url = artist_lists[i]
    request = requests.get(url)
    text = request.text
    bs = BeautifulSoup(text, 'html.parser')
    all_list = bs.select('#songlist .song a')
    for song_list in all_list:
        url = song_list.get('href')
        song_lists.append(url)
```

```python
song_lists
```

```python
all_url = []
for song_list in song_lists:
    req = requests.get(song_list)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    url = soup.select('#downloadbox a')
    for down_list in url:
        url = down_list.get('href')
        all_url.append(url)
```
