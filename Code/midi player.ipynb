```python
from music21 import midi
import pretty_midi
import os
import mido
from mido import MidiFile
import time
import rtmidi
import midi
from midi import read_midifile
```

```python
def playMidi(filename):
    mf = midi.MidiFile()
    mf.open(filename)
    mf.read()
    mf.close()
    s = midi.translate.midiFileToStream(mf)
    s.show('midi')
```

```python
def get_filenames(midifiles_dir):
    midi_root = midifiles_dir
    directories = []
    for filename in os.listdir(midi_root):
        if isinstance(filename, pretty_midi.PrettyMIDI):
            path = os.path.join(midi_root, filename)
            midis = pretty_midi.PrettyMIDI(path)
            directories.append(midis)
        else:
            try:
                path = os.path.join(midi_root, filename)
                midis = pretty_midi.PrettyMIDI(path)
                directories.append(path)
            except:
                continue
    return directories
```

```python
filename = get_filenames('./')
get_filenames('./')
```

```python
for i in range(len(filename)):
    print(filename[i])
    playMidi(filename[i])
```
